## Data Cleaning

### Filling missing values
*When to use mean or median to fill missing values*
* When to use mean: |Skew| < 0.5 cause there are no or very few outliers.
* when to use median: |Skew| > 0.5 cause if the number is very large, it could be beacause of outliers

```python
train_df.select_dtypes(include='number').skew()
```

```python
# Filling missing values using median and mean
for column in train_df.select_dtypes(include = 'number').columns:
    if train_df[column].skew() > 0.5:
        train_df[column] = train_df[column].fillna(train_df[column].median())
    else:
        train_df[column] = train_df[column].fillna(train_df[column].mean())
```

```python
train_df.select_dtypes(include = 'number').isna().sum()
```

```python
train_df.HomePlanet.mode()
```

```python
for column in train_df.select_dtypes(include = 'object').columns:
    train_df[column] = train_df[column].fillna(train_df[column].mode()[0]) # mode returns a series, that's why is used index 0
```

### Checking skewness and using boxcox to reduce the skewness

```python
# checking for skewness
import matplotlib.pyplot as plt
from scipy.stats import skew

# numerical columns
numeric_cols = train_df.select_dtypes(include=['int64', 'float64']).columns

# Identify highly skewed columns
skewed_cols = []
print("Highly Skewed Columns:\n")
for col in numeric_cols:
    skew_val = skew(train_df[col])
    if abs(skew_val) > 1:
        print(f"{col}  Skewness: {skew_val:.2f}")
        skewed_cols.append(col)

# Plot distributions using sns.displot
for col in skewed_cols:
    sns.displot(data=train_df, x=col, kde=True, height=4, bins=30, aspect=1.5)
    plt.title(f'Distribution of {col} (Skewness Visualized)')
    plt.xlabel(col)
    plt.grid(True)
    plt.show()
```

## Solving for skewness

```python
from scipy.stats import skew, boxcox

# Select numeric columns
numeric_cols = train_df.select_dtypes(include=['int64', 'float64']).columns

# Identify highly skewed columns
skewed_cols = []
for col in numeric_cols:
    skew_val = skew(train_df[col].dropna())
    if abs(skew_val) > 1:
        skewed_cols.append(col)

print("Skewed Columns:", skewed_cols)

# Apply Box-Cox to each skewed column
for col in skewed_cols:
    original_data = train_df[col].dropna()

    # Fix zero or negative values by shifting
    shift = 0
    if (original_data <= 0).any():
        shift = abs(original_data.min()) + 1
        shifted_data = original_data + shift
        print(f"Shifted {col} by +{shift} to apply Box-Cox")
    else:
        shifted_data = original_data

    # Box-Cox transformation (on shifted_data)
    transformed_data, _ = boxcox(shifted_data)

    # Calculate skewness values
    skew_before = skew(original_data)
    skew_after = skew(transformed_data)

    # Plot before
    g = sns.displot(original_data, bins=30, kde=True)
    g.fig.suptitle(f'{col} - Before Box-Cox (Skewness: {skew_before:.2f})', y=1.02)
    for ax in g.axes.flat:
        ax.grid(True)
    plt.show()

    # Plot after
    g = sns.displot(transformed_data, bins=30, kde=True)
    g.fig.suptitle(f'{col} - After Box-Cox (Skewness: {skew_after:.2f})', y=1.02)
    for ax in g.axes.flat:
        ax.grid(True)
    plt.show()

```
